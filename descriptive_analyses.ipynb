```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named df, and the genre columns are labeled as "action", "thriller", ..., "romance"
# Select the genre columns
genre_columns = ['Drama' ,'Comedy', 'Action',
 'Thriller' ,'Romance' ,'Adventure' ,'Crime' ,'Horror' ,'Science Fiction',
 'Family' ,'Fantasy', 'Mystery' ,'Animation' ,'History' ,'Music' ,'War',
 'Documentary', 'Western' ,'TV Movie']

# Select genre columns and aggregate counts
genre_counts = df[genre_columns].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust size as needed
genre_counts.plot(kind='bar', color='skyblue')

plt.title('Counts of Different Movie Genres')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```

```python
actors=df[df['actor_previous_success']<50]
plt.figure(figsize=(10, 6))  # Adjust size as needed
plt.hist(actors['actor_previous_success'], bins=10, color='skyblue', edgecolor='black')  # Adjust number of bins as needed
plt.title('Actor Previous Success (Excluding 56 instances above 50)')
plt.xlabel('Average of Revenue/Budget Ratio for Actors previous releases')
plt.ylabel('Frequency')
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named df, and the genre columns are labeled as "action", "thriller", ..., "romance"
# Select the genre columns
Ratings = ['age_rating_NC-17', 'age_rating_NR',
 'age_rating_PG', 'age_rating_PG-13', 'age_rating_R',
 'age_rating_unrated']

# Select genre columns and aggregate counts
Ratings = df[Ratings].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust size as needed
Ratings.plot(kind='bar', color='skyblue')

plt.title('Counts of Different Movie Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```
